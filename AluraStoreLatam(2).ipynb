### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingreso1 = tienda['Precio'].sum()
ingreso2 = tienda2['Precio'].sum()
ingreso3 = tienda3['Precio'].sum()
ingreso4 = tienda4['Precio'].sum()
print(f"Ingreso total Tienda 1: ${ingreso1:,.2f}")
print(f"Ingreso total Tienda 2: ${ingreso2:,.2f}")
print(f"Ingreso total Tienda 3: ${ingreso3:,.2f}")
print(f"Ingreso total Tienda 4: ${ingreso4:,.2f}")
```

# 2. Ventas por categoría

```python
ventas_categoria_1 = tienda.groupby('Categoría del Producto')['Precio'].sum()
ventas_categoria_2 = tienda2.groupby('Categoría del Producto')['Precio'].sum()
ventas_categoria_3 = tienda3.groupby('Categoría del Producto')['Precio'].sum()
ventas_categoria_4 = tienda4.groupby('Categoría del Producto')['Precio'].sum()

print("Ventas por categoría - Tienda 1:")
for categoria, total in ventas_categoria_1.items():
    print(f"{categoria}: ${total:,.2f}")

print("\nVentas por categoría - Tienda 2:")
for categoria, total in ventas_categoria_2.items():
    print(f"{categoria}: ${total:,.2f}")

print("\nVentas por categoría - Tienda 3:")
for categoria, total in ventas_categoria_3.items():
    print(f"{categoria}: ${total:,.2f}")

print("\nVentas por categoría - Tienda 4:")
for categoria, total in ventas_categoria_4.items():
    print(f"{categoria}: ${total:,.2f}")
```

# 3. Calificación promedio de la tienda


```python
calif_prom_1 = tienda['Calificación'].mean()
calif_prom_2 = tienda2['Calificación'].mean()
calif_prom_3 = tienda3['Calificación'].mean()
calif_prom_4 = tienda4['Calificación'].mean()

print(f"Calificación promedio Tienda 1: {calif_prom_1:.2f}")
print(f"Calificación promedio Tienda 2: {calif_prom_2:.2f}")
print(f"Calificación promedio Tienda 3: {calif_prom_3:.2f}")
print(f"Calificación promedio Tienda 4: {calif_prom_4:.2f}")
```

# 4. Productos más y menos vendidos

```python
producto_top_1 = tienda['Producto'].value_counts().idxmax()
producto_top_2 = tienda2['Producto'].value_counts().idxmax()
producto_top_3 = tienda3['Producto'].value_counts().idxmax()
producto_top_4 = tienda4['Producto'].value_counts().idxmax()

producto_low_1 = tienda['Producto'].value_counts().idxmin()
producto_low_2 = tienda2['Producto'].value_counts().idxmin()
producto_low_3 = tienda3['Producto'].value_counts().idxmin()
producto_low_4 = tienda4['Producto'].value_counts().idxmin()

print(f"Producto más vendido Tienda 1: {producto_top_1}")
print(f"Producto menos vendido Tienda 1: {producto_low_1}")
print(f"Producto más vendido Tienda 2: {producto_top_2}")
print(f"Producto menos vendido Tienda 2: {producto_low_2}")
print(f"Producto más vendido Tienda 3: {producto_top_3}")
print(f"Producto menos vendido Tienda 3: {producto_low_3}")
print(f"Producto más vendido Tienda 4: {producto_top_4}")
print(f"Producto menos vendido Tienda 4: {producto_low_4}")
```

# 5. Envío promedio por tienda

```python
envio_prom_1 = tienda['Costo de envío'].mean()
envio_prom_2 = tienda2['Costo de envío'].mean()
envio_prom_3 = tienda3['Costo de envío'].mean()
envio_prom_4 = tienda4['Costo de envío'].mean()

print(f"Costo promedio de envío Tienda 1: {envio_prom_1:.2f}")
print(f"Costo promedio de envío Tienda 2: {envio_prom_2:.2f}")
print(f"Costo promedio de envío Tienda 3: {envio_prom_3:.2f}")
print(f"Costo promedio de envío Tienda 4: {envio_prom_4:.2f}")
```

```python
ingresos_totales = {
    'Tienda 1': ingreso1,
    'Tienda 2': ingreso2,
    'Tienda 3': ingreso3,
    'Tienda 4': ingreso4
}

plt.figure(figsize=(8, 6))
plt.bar(ingresos_totales.keys(), ingresos_totales.values(), color=['blue', 'green', 'red', 'purple'])
plt.title('Ingresos Totales por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Ingresos ($)')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

# 2. Gráfico de torta: categorías Tienda 1
colors = plt.cm.Paired.colors[:len(ventas_categoria_1)]
plt.figure(figsize=(12, 10))
ventas_categoria_1.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=colors, textprops={'fontsize': 12})
plt.title('Distribución de Ventas por Categoría - Tienda 1', fontsize=14)
plt.ylabel('')
plt.tight_layout()
plt.show()

# 3. Calificaciones promedio
calificaciones_promedio = {
    'Tienda 1': calif_prom_1,
    'Tienda 2': calif_prom_2,
    'Tienda 3': calif_prom_3,
    'Tienda 4': calif_prom_4
}

plt.figure(figsize=(8, 6))
plt.bar(calificaciones_promedio.keys(), calificaciones_promedio.values(), color=['cyan', 'magenta', 'yellow', 'gray'])
plt.title('Calificación Promedio por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Calificación Promedio')
plt.ylim(0, 5)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

El propósito de este análisis es determinar a qué tienda debe vender el Sr. Juan, considerando factores clave como los ingresos totales, ventas por categoría de productos, calificaciones promedio de clientes, productos más y menos vendidos, y costos promedio de envío.

Las categorías con mayores ventas son Electrónicos y Electrodomésticos en todas las tiendas, mientras que categorías como Libros y Juguetes presentan menores ingresos.

Considerando todos los factores analizados, la recomendación para el Sr. Juan es vender en **la Tienda 1**.

Esta tienda presenta los mayores ingresos totales y una fuerte presencia en las categorías más rentables (Electrónicos y Electrodomésticos). A pesar de tener un costo de envío promedio un poco más alto y una calificación ligeramente inferior, los beneficios económicos compensan estas diferencias.

En resumen, la elección dependerá de si el Sr. Juan prioriza ingresos o experiencia del cliente. Para maximizar ingresos, Tienda 1; para mejor valoración y envío económico, Tienda 3.
